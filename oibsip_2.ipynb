```python
def clean_carname(x):
    return x.split(" ")[0]

data.CarName.apply(clean_carname).value_counts()
```

```python
# Some issues are there in carname: toyota : toyouta , nissan : Nissan , maxda : maxda , volkswagen : vokswagen : vw , porsche : porcshce 
```

```python
def clean_again(x):
    name = x.split(" ")[0]
    if name == 'Nissan':
        return "nissan"
    elif name == 'toyouta':
        return "toyota"
    elif name == "maxda":
        return "mazda"
    elif name == "vokswagen" or name == "vw":
        return "volkswagen"
    elif name == "porcshce":
        return "porsche"
    else:
        return name
    
data.CarName = data.CarName.apply(clean_again)
    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor , GradientBoostingRegressor , AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
x = data.drop('price' , axis = 1)
y = data.price
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y, test_size = 0.2 ,random_state = 42)
```

```python
xtrain.shape , xtest.shape
```

```python
xtrain
```

```python
one = OneHotEncoder()
one.fit(x[['CarName' , 'fueltype' , 'aspiration' , 'doornumber' , 'carbody' , 'drivewheel' , 'enginetype' , 'cylindernumber', 'fuelsystem']])
categories = one.categories_
categories
```

```python
col1 = ColumnTransformer(transformers = [('onehotencoder' , OneHotEncoder(categories = categories , drop = 'first') , 
                                        [0,1,2,3,4,5,10,11,13])],
                        remainder = 'passthrough')
```

```python
xtrain = col1.fit_transform(xtrain).todense()
xtest = col1.transform(xtest).todense()
```

```python
xtrain.shape
```

```python
col2 = ColumnTransformer(transformers = [('scaling' , StandardScaler() , slice(0,59))] , remainder = "passthrough")
```

```python
xtrain = col2.fit_transform(np.array(xtrain))
xtest = col2.transform(np.array(xtest))
```

```python
rf = RandomForestRegressor()
rf.fit(xtrain , ytrain)
ypred1 = rf.predict(xtest)
r2_score(ytest , ypred1)
```

```python
lr = LinearRegression()
lr.fit(xtrain , ytrain)
ypred2 = lr.predict(xtest)
r2_score(ytest , ypred2)
```

```python
gd = GradientBoostingRegressor()
gd.fit(xtrain , ytrain)
ypred3 = gd.predict(xtest)
r2_score(ytest, ypred3)
```

```python
ad = AdaBoostRegressor()
ad.fit(xtrain , ytrain)
ypred4 = ad.predict(xtest)
r2_score(ytest , ypred4)
```

```python
sv = SVR()
sv.fit(xtrain , ytrain)
ypred5 = sv.predict(xtest)
r2_score(ytest , ypred5)
```

```python
dt = DecisionTreeRegressor()
dt.fit(xtrain , ytrain)
ypred6 = dt.predict(xtest)
r2_score(ytest , ypred6)
```

```python
# Random Forest regressor is giving the best accuracy although we can do hyperparamter tuning using grid search cv but again data is to too less to capture the relationships.
```
